```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json 
import os
import itertools
from wordcloud import WordCloud
```

```python
data = pd.read_csv(r"phishing_site_urls.csv")  #Read the dataset 
```

```python
datac=data.copy() #Made a copy of the data to experiment so that we dont harm the actual dataset
```

```python
from nltk.tokenize import RegexpTokenizer 
tokenizer = RegexpTokenizer(r'[A-Za-z]+')
datac['Url_tokenized'] = datac.URL.map(lambda t: tokenizer.tokenize(t))
```

```python
def count_https(url):
    return url.count('https')

def count_http(url):
    return url.count('http')

def count_com(url):
    return url.count('com')

def count_dot(url):
    return url.count('.')

def count_slash(url):
    return url.count('/')

def count_question(url):
    return url.count('?')

def count_equals(url):
    return url.count('=')

def count_underscore(url):
    return url.count('_')

def count_ampersand(url):
    return url.count('&')

def count_percent(url):
    return url.count('%');

def count_orgwiki(url):
    return url.count('org wiki');

def count_html(url):
    return url.count('html');

def count_indexphp(url):
    return url.count('index php');

def count_battlenet(url):
    return url.count('battle net');
```

```python
datac['count_https']=datac['URL'].apply(lambda i : count_https(i))
datac['count_http']=datac['URL'].apply(lambda i : count_http(i))
datac['count_com']=datac['URL'].apply(lambda i : count_com(i))
datac['count_dot']=datac['URL'].apply(lambda i : count_dot(i))
datac['count_slash']=datac['URL'].apply(lambda i : count_slash(i))
datac['count_question']=datac['URL'].apply(lambda i : count_question(i))
datac['count_equals']=datac['URL'].apply(lambda i : count_equals(i))
datac['count_underscore']=datac['URL'].apply(lambda i : count_underscore(i))
datac['count_ampersand']=datac['URL'].apply(lambda i : count_ampersand(i))
datac['count_percent']=datac['URL'].apply(lambda i : count_percent(i))
datac['count_orgwiki']=datac['URL'].apply(lambda i : count_orgwiki(i))
datac['count_html']=datac['URL'].apply(lambda i : count_html(i))
datac['count_indexphp']=datac['URL'].apply(lambda i : count_indexphp(i))
datac['count_battlenet']=datac['URL'].apply(lambda i : count_battlenet(i))
```

```python
datac.drop(['URL'],axis=1,inplace=True)
datac.head()
```

```python
datac.drop(['Url_tokenized'],axis=1,inplace=True)
datac.head()
```

```python
from sklearn.model_selection import train_test_split

x = datac.iloc[:, 1:]
y = datac.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.1,random_state= 69)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
datac['Label']= label_encoder.fit_transform(datac['Label'])
datac['Label'].unique()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_classifier = KNeighborsClassifier()
knn_classifier.fit(x_train, y_train)
knn_classifier.score(x_test,y_test)
```
